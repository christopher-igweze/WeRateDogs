# Assesing the data

## Quality

1. the 'id' in df_json should be 'tweet_id'
2. timestamp is given as object instead of datetime
3. none is used to represent null dog personalities
4. in the 'rating denominator' column, some values are greater than 10 while some are less than 10
5. name column has none instead of nan
6. expanded_urls has null values
7. Extract ratings from tweets
8. we need to remove retweets (i.e rows with retweet id's)

## Tidiness

1. in df1 the dog type should be collapsed in one column
2. in df_json all we need is id, retweet_count, favorite_count
3. all datasets should be combined

## Cleaning

Creating a new copy

```python
df1 = df.copy()
df2 = df1.copy()
df_json = df2.copy()
```

Define: I am going to change the 'id' in df_json to 'tweet_id'

```python
# no 1
df_json.rename(columns = {'id' : 'tweet_id'}, inplace=True)
```

```python
#test
df_json.columns
```

Define: I am going to correct timestamp which is given as object instead of datetime

```python
# no 2
df1.timestamp = pd.to_datetime(df1.timestamp)
```

```python
#test
df1.timestamp.dtype
```

Define: I am going to replace none with np.nan for the dog types

```python
# no 3
cols = ['pupper', 'puppo', 'doggo', 'floofer']

for i in cols:
    df1[i] = df1[i].replace('None', np.nan)
```

```python
#test
df1.pupper
```

Define: I am going to correct the 'rating denominator' column, where some values are greater than 10 while some are less than 10

```python
# no 4
df1 = df1[df1['rating_denominator'] == 10]
```

```python
#test
df1.rating_denominator.unique()
```

Define: I am going to replace none with np.nan in the name column

```python
# no 5
df1[~df1.name.str.istitle()].name.unique()
```

```python
#test
df1.name.unique()
```

Define: I am  going to remove null values in expanded_urls

```python
# no 6
df1.expanded_urls.dropna()
```

```python
#test
#df1.expanded_urls.isna()
df1
```

Define: I am going to extract ratings from tweets

```python
# no 7
df1[df1.text.str.contains(r"(\d+\.\d*\/\d+)")][['text', 'rating_numerator']]

new_ratings = df1[df1.text.str.contains(r"(\d+\.\d*\/\d+)")]['text'].str.extract(r"(\d+\.\d*(?=\/\d+))")

df1.loc[new_ratings.index, 'rating_numerator'] = new_ratings.values

df1.rating_numerator = df1.rating_numerator.astype('float')
```

```python
#test
df.rating_numerator.sample()
```

Define: I am going to remove retweets (i.e rows with retweet id's)

```python
# no 8
df1 = df1.query('in_reply_to_status_id == "NaN" &\
                             in_reply_to_user_id == "NaN" &\
                             retweeted_status_id == "NaN" &\
                             retweeted_status_user_id == "NaN"')

cols = ['in_reply_to_status_id','in_reply_to_user_id','retweeted_status_id',
           'retweeted_status_user_id', 'retweeted_status_timestamp']
df1.drop(columns = cols, axis=1, inplace=True)
```

```python
#test
#df1['in_reply_to_user_id'].unique()
df1.columns
```

#### Tidy

Define: I am going to collapse the dog type to a single column

```python
# no 1
cols = ['doggo', 'floofer', 'pupper', 'puppo']

df1['dog_stage'] = df1[cols].apply(lambda x: ', '.join(x.dropna().astype(str)),axis =1)

df1.dog_stage = df1.dog_stage.replace('', np.nan).astype('category')

df1 = df1.drop(columns = cols, axis =1)
```

```python
#test
df1.dog_stage
```

Define: I am going to remove all columns leaving only, id, retweet_count, favorite_count

```python
# no 2
cols = ['tweet_id', 'retweet_count', 'favorite_count']

df_json = df_json[cols]
```

```python
df_json.columns
```

```python
df_json
```

Define: I am going to merge all datasets

```python
# no 3
df = pd.merge(df1, df2, on='tweet_id')
df = pd.merge(df, df_json, on='tweet_id')
```

## Visualization

```python
twitter_master = pd.read_csv('twitter_master.csv')

twitter_master = df
```

1. Distribution of ratings based on frequency

```python
data = twitter_master.rating_numerator.value_counts()

x = data.index
y = data.values
fig, ax = plt.subplots(figsize=(12, 6))
g = sns.barplot(x, y, palette='Blues_d', ax=ax)
ax.set(xlabel='Ratings', ylabel='Frequency', title='Ratings frequency')
plt.show()
```

From the plot, most tweets have a rating of either 10, 11, 12, or 13

2. Dog stage comparison

```python
data = twitter_master.groupby('dog_stage').count()['tweet_id']
ax = sns.barplot(y=data.index, x=data.values, palette='Blues_d')
ax.set(xlabel='Count', ylabel='Dog stage', title='Dog Stage Counts')
plt.show()
```

From this plot, the most common stage is pupper

3. Now we would like to compare the tweets by their sources

```python
data = twitter_master.groupby('source').count()['tweet_id']
ax = sns.barplot(y=data.index, x=data.values, palette='Blues_d')
ax.set(xlabel='Count', ylabel='Tweet source', title='Tweet Source Counts')
plt.show()
```

```python
As we can see, majority of these posts were by Iphones.
```
